```python
import pandas as pd
import numpy as np

# Read datasets
cpi = pd.read_csv('./inflationIndicatorsUpdates/cpi_data.csv')
cpi_base = pd.read_csv('./inflationIndicatorsUpdates/CPI.csv')
IR = pd.read_csv('./inflationIndicatorsUpdates/InterestRate.csv')
gdp = pd.read_csv('./inflationIndicatorsUpdates/GDP.csv')
eu = pd.read_csv('./inflationIndicatorsUpdates/EUR_USD.csv')
Ir = pd.read_csv('./inflationIndicatorsUpdates/RealInterestRate.csv')
er_usd = pd.read_csv('./inflationIndicatorsUpdates/ExchangeRate_usd.csv')
er_eur = pd.read_csv('./inflationIndicatorsUpdates/ExchangeRate_eur.csv')
unem = pd.read_csv('./inflationIndicatorsUpdates/UnemploymentRate.csv')
xau = pd.read_csv('./inflationIndicatorsUpdates/XAU_me.csv')

# Convert 'year', 'month', and 'day' columns to integers
er_usd['year'] = er_usd['year'].astype(int)
er_usd['month'] = er_usd['month'].astype(int)
er_usd['day'] = er_usd['day'].astype(int)

eu['year'] = eu['year'].astype(int)
eu['month'] = eu['month'].astype(int)
eu['day'] = eu['day'].astype(int)

er_eur['year'] = er_eur['year'].astype(int)
er_eur['month'] = er_eur['month'].astype(int)
er_eur['day'] = er_eur['day'].astype(int)


cpi['year'] = cpi['year'].astype(int)
cpi_base['year'] = cpi_base['year'].astype(int)
Ir['year'] = Ir['year'].astype(int)
xau['year'] = xau['year'].astype(int)
xau['month'] = xau['month'].astype(int)
xau['day'] = xau['day'].astype(int)

IR['year'] = IR['year'].astype(int)

gdp['year'] = gdp['year'].astype(int)

unem['year'] = unem['year'].astype(int)

er_usd.set_index(['year', 'month', 'day'], inplace=True)

# Merge datasets based on year, month, and day from er_usd DataFrame
InflationIndicator_df = (
    pd
    .merge(er_usd, er_eur, on=['year', 'month', 'day'], how='left', suffixes=('_usd', '_eur'))
    .merge(eu, on=['year', 'month', 'day'], how='left', suffixes=('_usd', '_eu'))
    .merge(xau,on=['year', 'month', 'day'], how='left', suffixes=('_usd', '_xau'))
    .merge(cpi, on=['year'], how='left', suffixes=('_usd', '_cpi'))
    .merge(cpi_base, on=['year'], how='left', suffixes=('_usd', '_cpi'))
    .merge(Ir, on=['year'], how='left', suffixes=('_usd', '_ir'))
    .merge(IR, on=['year'], how='left', suffixes=('_usd', '_ir'))
    .merge(gdp, on=['year'], how='left', suffixes=('_usd', '_gdp'))
    .merge(unem, on=['year'], how='left', suffixes=('_usd', '_unem'))
)


# Optionally, sort the combined DataFrame by the 'year', 'month', and 'day' columns
InflationIndicator_df.sort_values(['year', 'month', 'day'], inplace=True)





# Display the first few rows of the combined DataFrame
print(InflationIndicator_df.head(60))
```
